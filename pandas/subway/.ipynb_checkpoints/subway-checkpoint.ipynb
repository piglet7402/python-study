```python
new_title1 = []
```

```python
for per_time in range(0, 24):
    bef = per_time
    aft = per_time + 1
    timer = "{}시~{}시 승하차인원".format(bef, aft)
    new_title1.append(timer)
```

```python
new_title1
```

```python
for i in range(0, 4):
    a = list()
    b = new_title1.pop(0)
    a.append(b)
    new_title1.append(a[0])
```

```python
new_title1
```

```python
new_title1 = pd.DataFrame(new_title1)
```

```python
new_title1 
```

### 시간별 승하차 인원 합치기

```python
time = 4
```

```python
for user_sum in range(3, 42, 2):
    bef_time = time
    aft_time = time + 1
    
    bef_col = user_sum
    aft_col = user_sum + 1
    
    timeb_a = station1.iloc[:, [bef_col, aft_col]]
    station1['{}~{}시 이용객 합계'.format(bef_time, aft_time)] = timeb_a.sum(axis=1)
    ++time
```

```python
station1
```

```python
timeb_a = station1.iloc[:, [3, 4]]
```

```python
station1['{}~{}시 이용객 합계'.format(4, 5)] = timeb_a.sum(axis=1)
```

```python
station1
```
